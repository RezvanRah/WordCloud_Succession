Based on the script of the season 4 finale, 'With Open Eyes,' written by Jesse Armstrong, this word cloud highlights the most frequently used words in the episode. A unique and personalized keepsake for any fan of the show.</h3>

```python
# Importing neccessary libraries 
import nltk
from nltk.corpus import stopwords
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# Loading the text file
with open('Succession_final_ep.txt', 'r') as file:
    text = file.read()

# Set the stopwords
stop_words = set(stopwords.words('english'))
new_words = ['okay','oh', 'uh', 'huh', 'go', 'got', 'yeah', 'hey', 'cause', 'know', 
             'like', "i'm", 'well', "we're", 'said', 'laugh', 'see', "can't", 'um', "i'll", 'phone', 'ringing']
stop_words.update(new_words)


# Generating the word cloud
wc = WordCloud(width = 800, height = 800, background_color='black', colormap='viridis', stopwords=stop_words)

wc.generate(text)
```

```python
# Display the word cloud
plt.figure(figsize=(8, 8))
plt.imshow(wc)
plt.axis("off")
plt.show()
```

```python
# Save to a file
wc.to_file('WordCloud.png')
```
